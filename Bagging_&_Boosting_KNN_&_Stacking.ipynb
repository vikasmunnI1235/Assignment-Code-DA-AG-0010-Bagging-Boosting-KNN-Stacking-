Question 1 : What is the fundamental idea behind ensemble techniques? How does
bagging differ from boosting in terms of approach and objective?

ANS
Bagging (Bootstrap Aggregating)
Approach: Bagging involves training multiple instances of the same learning algorithm on different subsets of the training data. Each subset is created by bootstrapping (sampling with replacement) from the original dataset. Each model is trained independently, and their predictions are then aggregated (e.g., by averaging for regression or majority voting for classification).
Objective: The primary objective of bagging is to reduce variance in high-variance, low-bias models (like decision trees). By training on different subsets, the models become less sensitive to the specific training data, thus reducing the variability of the final prediction.
Examples: Random Forest is a prominent example of bagging, where multiple decision trees are trained on bootstrapped samples, and predictions are averaged.

Boosting
Approach: Boosting also combines multiple models, but it does so sequentially and adaptively. It starts by training an initial model on the entire dataset. Subsequent models are trained to correct the errors of the previous models. More weight is given to misclassified or poorly predicted instances, forcing the new models to focus on these difficult examples.
Objective: The main objective of boosting is to reduce bias and to some extent variance, by converting a set of weak learners into a strong learner. It iteratively improves the model by focusing on errors made in previous iterations.
Examples: AdaBoost, Gradient Boosting Machines (GBM), and XGBoost are popular boosting algorithms.


Question 2: Explain how the Random Forest Classifier reduces overfitting compared to
a single decision tree. Mention the role of two key hyperparameters in this process.

ANS

A Random Forest Classifier reduces overfitting by combining predictions from many decision trees instead of relying on just one. Each tree is trained on different random samples and features, which lowers variance and improves generalization.

Two key hyperparameters:

n_estimators – Number of trees in the forest; more trees reduce overfitting by averaging results.

max_depth (or max_features) – Limits tree complexity or features used, preventing trees from becoming too specific to training data.

KNN Classifier – Wine Dataset

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# 1. Load dataset
data = load_wine()
X, y = data.data, data.target

# 2. Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 3. KNN without scaling
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Without Scaling")
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 4. Apply StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=5)
knn_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = knn_scaled.predict(X_test_scaled)

print("With Scaling")
print("Accuracy:", accuracy_score(y_test, y_pred_scaled))
print(classification_report(y_test, y_pred_scaled))

# 5. GridSearchCV for best K and distance metric
param_grid = {
    'n_neighbors': range(1, 21),
    'metric': ['euclidean', 'manhattan']
}

grid = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid.fit(X_train_scaled, y_train)

print("Best Parameters:", grid.best_params_)
print("Best CV Score:", grid.best_score_)
```

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report

# Load data
X, y = load_wine(return_X_y=True)

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# 1. KNN without scaling
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print("Unscaled Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

# 2. Apply StandardScaler
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)

knn.fit(X_train_s, y_train)
y_pred_s = knn.predict(X_test_s)

print("Scaled Accuracy:", accuracy_score(y_test, y_pred_s))
print(classification_report(y_test, y_pred_s))

# 3. GridSearchCV
param_grid = {
    "n_neighbors": range(1, 21),
    "metric": ["euclidean", "manhattan"]
}

grid = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5
)

grid.fit(X_train_s, y_train)

print("Best Parameters:", grid.best_params_)

# 4. Optimized KNN
best_knn = grid.best_estimator_
y_pred_opt = best_knn.predict(X_test_s)

print("Optimized Accuracy:", accuracy_score(y_test, y_pred_opt))
print(classification_report(y_test, y_pred_opt))
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

X, y = load_breast_cancer(return_X_y=True)

# Scaling
X = StandardScaler().fit_transform(X)

# PCA
pca = PCA()
X_pca = pca.fit_transform(X)

# Scree plot
plt.plot(pca.explained_variance_ratio_.cumsum())
plt.xlabel("Components")
plt.ylabel("Cumulative Variance")
plt.show()

# Retain 95% variance
pca = PCA(n_components=0.95)
X_reduced = pca.fit_transform(X)

# Train/Test
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3)

knn = KNeighborsClassifier()
knn.fit(Xtr, ytr)
print("Original Accuracy:", accuracy_score(yte, knn.predict(Xte)))

knn.fit(X_reduced[:len(Xtr)], ytr)
print("PCA Accuracy:", accuracy_score(yte, knn.predict(X_reduced[len(Xtr):])))
```

```python
from sklearn.datasets import make_regression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt

X, y = make_regression(n_samples=500, n_features=10, noise=20)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.3)

# Distance comparison
for metric in ['euclidean', 'manhattan']:
    knn = KNeighborsRegressor(n_neighbors=5, metric=metric)
    knn.fit(Xtr, ytr)
    print(metric, "MSE:", mean_squared_error(yte, knn.predict(Xte)))

# K vs MSE
ks = [1, 5, 10, 20, 50]
mse = []

for k in ks:
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(Xtr, ytr)
    mse.append(mean_squared_error(yte, knn.predict(Xte)))

plt.plot(ks, mse, marker='o')
plt.xlabel("K")
plt.ylabel("MSE")
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

# 1. Load the Breast Cancer dataset
X, y = load_breast_cancer(return_X_y=True)

# Scale the data first for consistent PCA and KNN
X_scaled = StandardScaler().fit_transform(X)

# 2. Apply PCA and plot the scree plot (explained variance ratio).
pca_full = PCA()
pca_full.fit(X_scaled) # Fit PCA on scaled data to get all components

# Scree plot
plt.figure(figsize=(8, 5))
plt.plot(range(1, len(pca_full.explained_variance_ratio_) + 1), pca_full.explained_variance_ratio_.cumsum(), marker='o', linestyle='--')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance Ratio")
plt.title("Scree Plot (Explained Variance)")
plt.grid(True)
plt.show()

# 3. Retain 95% variance and transform the dataset.
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_scaled)
print(f"Number of components retained for 95% variance: {X_pca_95.shape[1]}")

# 4. Train KNN on the original data (scaled) and PCA-transformed data (95% variance), then compare accuracy.
# Split the scaled original data and the 95%-variance PCA data consistently
X_train_orig, X_test_orig, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
X_train_pca_95, X_test_pca_95, _, _ = train_test_split(X_pca_95, y, test_size=0.3, random_state=42) # Using 'y' directly since it's the same target for both splits

# KNN on original (scaled) data
knn_orig = KNeighborsClassifier()
knn_orig.fit(X_train_orig, y_train)
print("\nKNN on Original (Scaled) Data Accuracy:", accuracy_score(y_test, knn_orig.predict(X_test_orig)))

# KNN on PCA-transformed (95% variance) data
knn_pca = KNeighborsClassifier()
knn_pca.fit(X_train_pca_95, y_train)
print("KNN on PCA (95% variance) Data Accuracy:", accuracy_score(y_test, knn_pca.predict(X_test_pca_95)))

# 5. Visualize the first two principal components using a scatter plot (color by class).
# Use PCA specifically for visualization with 2 components to ensure there are always two components to plot.
pca_viz = PCA(n_components=2)
X_pca_viz = pca_viz.fit_transform(X_scaled)

plt.figure(figsize=(10, 7))
plt.scatter(X_pca_viz[:, 0], X_pca_viz[:, 1], c=y, cmap='viridis', alpha=0.8)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("First Two Principal Components of Breast Cancer Dataset")
plt.colorbar(label='Class')
plt.grid(True)
plt.show()
```

KNN + Imputation + Trees (Pima Dataset)

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from time import time

data = pd.read_csv("pima_diabetes.csv")
X = data.drop("Outcome", axis=1)
y = data["Outcome"]

# Imputation
imputer = KNNImputer()
X = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

methods = ['brute', 'kd_tree', 'ball_tree']
for m in methods:
    knn = KNeighborsClassifier(algorithm=m)
    start = time()
    knn.fit(X_train, y_train)
    print(m, "Accuracy:", knn.score(X_test, y_test), "Time:", time()-start)
```

```python
import pandas as pd
from sklearn.impute import KNNImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from time import time
import requests
import numpy as np # Import numpy for np.nan

# Download the dataset if it's not already present
file_name = "pima_diabetes.csv"
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.csv"
try:
    with open(file_name, 'x') as f:
        f.write(requests.get(url).text)
except FileExistsError:
    pass

# Define column names as the dataset does not have a header
column_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
data = pd.read_csv(file_name, names=column_names)

# The Pima Indians Diabetes dataset often uses 0 to denote missing values for certain columns
# Replacing 0s with NaN for columns where 0 is not a valid physical measurement
# Use np.nan for compatibility with scikit-learn imputers
data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = data[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0, np.nan)
X = data.drop("Outcome", axis=1)
y = data["Outcome"]

# Imputation
imputer = KNNImputer()
X = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # Added random_state for reproducibility

methods = ['brute', 'kd_tree', 'ball_tree']
for m in methods:
    knn = KNeighborsClassifier(algorithm=m)
    start = time()
    knn.fit(X_train, y_train)
    print(f"{m} Accuracy: {knn.score(X_test, y_test):.4f} Time: {time()-start:.4f}s")
```
