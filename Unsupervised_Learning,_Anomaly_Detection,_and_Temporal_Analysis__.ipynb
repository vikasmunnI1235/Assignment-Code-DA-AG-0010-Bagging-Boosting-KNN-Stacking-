Question 3: What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

ANS

Clustering in unsupervised learning is the process of grouping data points into clusters such that points in the same cluster are more similar to each other than to those in other clusters.

Three popular clustering algorithms:

K-Means – Groups data into K clusters based on distance from cluster centroids.

Hierarchical Clustering – Builds a tree-like structure (dendrogram) of clusters.

DBSCAN – Forms clusters based on data density and identifies noise/outliers.


Question 4: Explain the concept of anomaly detection and its significance.

ANS

Anomaly detection is the process of identifying data points, events, or observations that deviate significantly from the majority of the data. These unusual points are often called outliers or anomalies.

Significance in Machine Learning:

Fraud Detection – Identify unusual transactions in banking.

Network Security – Detect cyber attacks or unauthorized access.

Fault Detection – Spot defects or failures in machines or systems.

Data Quality – Find errors or inconsistencies in datasets.

Question 5: List and briefly describe three types of anomaly detection techniques.

ANS .

Three Types of Anomaly Detection Techniques

Statistical Methods

Assume data follows a known distribution (e.g., Gaussian).

Points that significantly deviate from the expected distribution are flagged as anomalies.

Example: Z-score method.

Machine Learning-Based Methods

Use models to learn normal patterns in data and identify deviations.

Can be supervised (labeled anomalies) or unsupervised (no labels).

Examples: Isolation Forest, One-Class SVM.

Distance or Density-Based Methods

Detect anomalies based on how far a point is from its neighbors or how sparse its region is.

Examples: DBSCAN, k-Nearest Neighbors (k-NN) anomaly detection.

Question 8: Write Python code to perform K-means clustering on a sample dataset.
(Include your Python code and output in the code box below.)

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Create a sample dataset with 2 features and 300 samples
X, y = make_blobs(n_samples=300, centers=3, cluster_std=0.6, random_state=42)

# Plot the dataset
plt.scatter(X[:, 0], X[:, 1], s=50, c='gray')
plt.title("Sample Data")
plt.show()

# Perform K-Means clustering with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Get cluster centers and labels
centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X')  # cluster centers
plt.title("K-Means Clustering")
plt.show()

# Print cluster centers
print("Cluster Centers:\n", centers)
```

Question 9: What is inheritance in OOP? Provide a simple example in Python.

```python
# Parent class
class Vehicle:
    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}")

# Child class inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, brand, model, doors):
        super().__init__(brand, model)  # Call parent constructor
        self.doors = doors

    def car_info(self):
        print(f"Brand: {self.brand}, Model: {self.model}, Doors: {self.doors}")

# Create an object of Car
my_car = Car("Toyota", "Corolla", 4)
my_car.display_info()   # Method inherited from Vehicle
my_car.car_info()       # Method of Car
```
